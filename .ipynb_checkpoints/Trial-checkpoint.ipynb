```python
import numpy as np
from astropy import modeling
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
import pandas as pd 

m = modeling.models.Gaussian1D(amplitude = 10, mean = 30, stddev = 5)
x = np.linspace(0,100,2000)
data = m(x)
plt.plot(x, data)
```

```python
#mixture of gaussians
m1 = modeling.models.Gaussian1D(amplitude = 8, mean = 40, stddev = 9)
data1 = m1(x)
final = data + data1
plt.plot(x, final)
plt.plot(x, data)
plt.plot(x, data1)
```

```python
#Plotting the experimental data 
table = pd.read_excel('Gaussian_test.xlsx')
table.head()
f = table['frequency']
a = f[0:1350]
intensity = table['intensity']
b = intensity[0:1350]
plt.plot(a, b)
```

```python
#Trying the curve_fit fucntion with gaussian with the fabricated gaussian mixture - because the experimnetal data did not work.
def gauss(x, a, x0, sigma):
    return a*np.exp(-(x-x0)**2/(2*sigma**2))
fig = plt.figure()
popt, pcov = curve_fit(gauss, x, final)
intensity1 = gauss(x, popt[0], popt[1], popt[2])
ax = fig.add_subplot(111)
#ax.plot(x, intensity1, c='k', label='Function')
ax.plot(x, intensity1, c = 'r',label='Best fit')
ax.scatter(x, final)
```
